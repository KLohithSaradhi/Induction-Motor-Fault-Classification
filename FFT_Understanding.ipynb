```python
from scipy.fft import rfft, rfftfreq
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sampling_rate = 500
duration = 2

NUMBER_OF_SAMPLES = duration * sampling_rate

frequency_of_signal = 2
frequency_of_noise = 10

x = np.linspace(0, duration, NUMBER_OF_SAMPLES, endpoint = False)
y = np.sin(2 * np.pi * frequency_of_signal * x) + 0.5 * np.sin(2 * np.pi * frequency_of_noise * x)

plt.plot(x,y)
```

```python
yf = rfft(y)
xf = rfftfreq(NUMBER_OF_SAMPLES, 1/sampling_rate)

plt.plot(xf, np.abs(yf))
```

```python
xf[np.argwhere(np.abs(yf) > 1)]
```
