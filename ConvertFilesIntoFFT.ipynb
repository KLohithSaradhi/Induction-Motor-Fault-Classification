```python
import os
import pandas as pd
from scipy.fft import rfft, rfftfreq
import numpy as np
```

```python
def convert(directory):
    
    for filename in os.listdir(directory):
    
        print(filename)
        data = pd.read_csv(os.path.join(directory, filename), skiprows = 20)
        ch1 = np.array(data.iloc[:,1]).astype(np.float32)
        ch2 = np.array(data.iloc[:,2]).astype(np.float32)
        ch3 = np.array(data.iloc[:,3]).astype(np.float32)
        ch4 = np.array(data.iloc[:,4]).astype(np.float32)

        Normalizing_factor = len(ch1)

        try:
            mag = np.abs(rfft(ch1)/Normalizing_factor)[::10]
            ch1_df.loc[len(ch1_df)] = [filename] + list(mag)

            mag = np.abs(rfft(ch2)/Normalizing_factor)[::10]
            ch2_df.loc[len(ch2_df)] = [filename] + list(mag)

            mag = np.abs(rfft(ch3)/Normalizing_factor)[::10]
            ch3_df.loc[len(ch3_df)] = [filename] + list(mag)

            mag = np.abs(rfft(ch4)/Normalizing_factor)[::10]
            ch4_df.loc[len(ch4_df)] = [filename] + list(mag)

        except:
            frequencies = rfftfreq(len(ch1), 1/10000)[::10]
            ch1_df = pd.DataFrame(columns=["ID"] + list(frequencies))
            ch2_df = pd.DataFrame(columns=["ID"] + list(frequencies))
            ch3_df = pd.DataFrame(columns=["ID"] + list(frequencies))
            ch4_df = pd.DataFrame(columns=["ID"] + list(frequencies))

            mag = np.abs(rfft(ch1)/Normalizing_factor)[::10]
            ch1_df.loc[len(ch1_df)] = [filename] + list(mag)

            mag = np.abs(rfft(ch2)/Normalizing_factor)[::10]
            ch2_df.loc[len(ch2_df)] = [filename] + list(mag)

            mag = np.abs(rfft(ch3)/Normalizing_factor)[::10]
            ch3_df.loc[len(ch3_df)] = [filename] + list(mag)

            mag = np.abs(rfft(ch4)/Normalizing_factor)[::10]
            ch4_df.loc[len(ch4_df)] = [filename] + list(mag)

    return ch1_df, ch2_df, ch3_df, ch4_df
```

```python
 for c in range(4,7):
    ch1_df, ch2_df, ch3_df, ch4_df = convert(str(c))

    ch1_df.to_csv(str(c) + "_ch1.csv", index = False)
    ch2_df.to_csv(str(c) + "_ch2.csv", index = False)
    ch3_df.to_csv(str(c) + "_ch3.csv", index = False)
    ch4_df.to_csv(str(c) + "_ch4.csv", index = False)
```
