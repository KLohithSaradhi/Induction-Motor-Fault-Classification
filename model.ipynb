```python
finite_X = finite_X.reset_index(drop=True)
finite_Y = finite_Y.reset_index(drop=True)
```

### Data Prep

```python
from sklearn.model_selection import train_test_split

clean_X = finite_X.drop(["0", "ID"], axis = 1)

clean_X = clean_X.reset_index(drop=True)

x_train, x_test, y_train, y_test = train_test_split(clean_X, finite_Y, test_size=0.2, random_state=0, stratify=finite_Y)
```

```python
x_train.shape
```

```python
np.max(x_train)
```

```python
n_train = np.reshape(np.linalg.norm(x_train, axis = 1), (-1, 1))
n_test = np.reshape(np.linalg.norm(x_test, axis = 1), (-1,1))
```

```python
normalized_x_train = (x_train / n_train) * 100
normalized_x_test = (x_test / n_test) * 100
```

```python
np.max(normalized_x_train)
```

```python
np.max(normalized_x_test)
```

```python
y_train = np.reshape(np.array(y_train), (-1,1))
y_test = np.reshape(np.array(y_test), (-1,1))
```

```python
from sklearn.preprocessing import OneHotEncoder

o = OneHotEncoder()
y_train_onehot = o.fit_transform(y_train).toarray()
y_test_onehot = o.transform(y_test).toarray()
```

### Model Building starts

```python
import tensorflow as tf

from keras.layers import Dense
from keras import Sequential
from keras.losses import SparseCategoricalCrossentropy, CategoricalCrossentropy
```

```python
model = Sequential([Dense(2500, activation = "relu"),
                    Dense(1200, activation = "relu"),
                    Dense(600, activation = "relu"),
                    Dense(300, activation = "relu"),
                    Dense(150, activation = "relu"),
                    Dense(75, activation = "relu"),
                    Dense(30, activation = "relu"),
                    Dense(7, activation = "softmax")])

model.compile(loss=SparseCategoricalCrossentropy(), optimizer = "adam",metrics=["accuracy"])
```

### PCA for reducing vector length

```python
from sklearn.decomposition import PCA

test_p = PCA()
test_p.fit(normalized_x_train)
```

```python
np.where(np.cumsum(test_p.explained_variance_ratio_) <= 0.9999999)
```

```python
p = PCA(n_components=1917)
p.fit(normalized_x_train)
```

```python
compressed_x_train = p.transform(normalized_x_train)
compressed_x_test = p.transform(normalized_x_test)
```

```python
compressed_x_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

newmodel = KNeighborsClassifier(n_neighbors=11)

newmodel.fit(compressed_x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(newmodel.predict(compressed_x_test), y_test)
```
