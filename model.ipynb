```python
import pandas as pd
import numpy as np
```

```python
ch1 = pd.read_csv("CH1.csv")
```

```python
ch1.info()
```

```python
X = ch1.select_dtypes(exclude = "int64")
Y = ch1["label"]
```

### Data Analysis

#### Labels Histogram

```python
pd.value_counts(Y)
```

```python
fft = X.select_dtypes("float64").T
```

##### Files with infinity components

```python
ch1.loc[np.where(np.max(fft) == np.inf)[0], "ID"]
```

```python
finite_X = X.drop(np.where(np.max(fft) == np.inf)[0])
finite_Y = Y.drop(np.where(np.max(fft) == np.inf)[0])
```

```python
finite_X.iloc[np.where(finite_X.isna().any(axis = 1))[0]]
```

```python
finite_Y = finite_Y.drop(np.where(finite_X.isna().any(axis = 1))[0])
finite_X = finite_X.drop(np.where(finite_X.isna().any(axis = 1))[0])
```

```python
finite_X = finite_X.reset_index(drop=True)
finite_Y = finite_Y.reset_index(drop=True)
```

```python
finite_X
```

### Data Prep

```python
from sklearn.model_selection import train_test_split

clean_X = finite_X.drop(["0", "ID"], axis = 1)

clean_X = clean_X.reset_index(drop=True)

x_train, x_test, y_train, y_test = train_test_split(clean_X, finite_Y, test_size=0.2, random_state=0, stratify=finite_Y)
```

```python
x_train.shape
```

```python
np.max(x_train)
```

```python
n_train = np.reshape(np.linalg.norm(x_train, axis = 1), (-1, 1))
n_test = np.reshape(np.linalg.norm(x_test, axis = 1), (-1,1))
```

```python
normalized_x_train = (x_train / n_train) * 100
normalized_x_test = (x_test / n_test) * 100
```

```python
np.max(normalized_x_train)
```

```python
np.max(normalized_x_test)
```

```python
y_train = np.reshape(np.array(y_train), (-1,1))
y_test = np.reshape(np.array(y_test), (-1,1))
```

```python
from sklearn.preprocessing import OneHotEncoder

o = OneHotEncoder()
y_train_onehot = o.fit_transform(y_train).toarray()
y_test_onehot = o.transform(y_test).toarray()
```

### Model Building starts

```python
import tensorflow as tf

from keras.layers import Dense
from keras import Sequential
from keras.losses import SparseCategoricalCrossentropy, CategoricalCrossentropy
```

```python
model = Sequential([Dense(2500, activation = "sigmoid"),
                   Dense(100, activation = "sigmoid"),
                   Dense(7, activation = "softmax")])

model.compile(loss=CategoricalCrossentropy(), optimizer = "rmsprop",metrics=["accuracy"])
```

```python
model.predict(x_test)
```

```python
newWeights = []

for l in model.get_weights():
    print(l.shape)
    try:
        x = l.shape[1]
        newWeights += [0.35  + 0.1 * np.random.randn(l.shape[0], x)]
    except:
        newWeights += [np.full(l.shape[0], 0.25)]
newWeights
```

```python
model.set_weights(newWeights)
model.weights
```

```python
model.predict(x_test)
```

```python
model.fit(normalized_x_train, y_train_onehot)
model.predict(x_test)
```
